```python
#Pour une bonne utilisation, on crée une fonction
def crawler(pagina):
  http = urllib3.PoolManager()
  try:
    pagina = http.request('GET', pagina)
  except:
    print('Page mot found', pagina)
  
  soup = BeautifulSoup(pagina.data, 'lxml')
  links = soup.find_all('a')
  #links = [str(link.get('href')) + ' - ' + str(link.contents) for link in links]  #pour retourner chaque lien avec son contenu
  links = [link.attrs for link in links]   #pour retourner tous les attributs d un lien, ça nous retourne les attributs tels href, title
  return links
```

```python
crawler('https://pt.wikipedia.org/wiki/Linguagem_de_programa%C3%A7%C3%A3o')
```
